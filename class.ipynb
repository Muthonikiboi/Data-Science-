```python
class Employee:
    empCount = 0
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.empCount += 1
        
    def displayCount(self):
        print("Total Employees:", Employee.empCount)
        
    def display(self):
        print("Name:", self.name , "Salary:", self.salary)
```

```python
    ob =Employee("Joy",1000000)
    ob.displayCount()
```
