```python
import math

def salesman_search(places_nodes, distances_btwn):
    # Initialize the solution.
    solution = [places_nodes[0]]
    remaining_nodes = places_nodes[1:]
    distance = 0

    # Iterate until all nodes are visited-going through the loop
    while remaining_nodes:
        current_node = solution[-1]
        min_distance = math.inf
        next_node = None

        # Find the nearest unvisited node--searches a shorter node
        for node in remaining_nodes:
            if distances_btwn[current_node][node] < min_distance:
                min_distance = distances_btwn[current_node][node]
                next_node = node

        # Add the nearest node to the solution.
        solution.append(next_node)
        remaining_nodes.remove(next_node)
        distance += min_distance

    # Add the distance from the last node to the starting node.
    distance += distances_btwn[solution[-1]][solution[0]]

    # Return the total distance.
    return distance


if __name__ == "__main__":
    # Initialization of all the nodes and distances.
    places_nodes = ["Nairobi Center", "Kirinyaga Town", "Nyeri Town", "Embu Town", "Meru Town", "Kisumu Town"]
    #distances for each node with all other nodes
    distances_btwn = {
        "Nairobi Center": {
            "Kirinyaga Town": 189,
            "Nyeri Town": 150.8,
            "Embu Town": 128.6,
            "Meru Town": 224.3,
            "Kisumu Town": 353
        },
        "Kirinyaga Town": {
            "Nairobi Center": 189,
            "Nyeri Town": 55.6,
            "Embu Town": 111.3,
            "Meru Town": 113.1,
            "Kisumu Town": 358
        },
        "Nyeri Town": {
            "Nairobi Center": 150.8,
            "Kirinyaga Town": 55.6,
            "Embu Town": 73.2,
            "Meru Town": 136.5,
            "Kisumu Town": 348.4
        },
        "Embu Town": {
            "Nairobi Center": 128.6,
            "Kirinyaga Town": 111.3,
            "Nyeri Town": 73.2,
            "Meru Town": 97.7,
            "Kisumu Town": 427.7
        },
        "Meru Town": {
            "Nairobi Center": 22.3,
            "Kirinyaga Town": 113.1,
            "Nyeri Town": 136.5,
            "Embu Town": 97.7,
            "Kisumu Town": 438.9
        },
        "Kisumu Town": {
            "Nairobi Center": 353,
            "Kirinyaga Town": 358,
            "Nyeri Town": 348.4,
            "Embu Town": 427.,
            "Meru Town": 438.9
        }
    }

    # Calculate the distances between the main node and other towns.
    main_node = "Nairobi Center"
    main_distances = distances_btwn[main_node]

    # Display the distances between each of the nodes 
    for node, distance in main_distances.items():
        print("Distance between the main node -", main_node, "and the other node -", node, "is", distance, "km")

    # graph search to solve the salesman search problem.
    distance = salesman_search(places_nodes, distances_btwn)

    # ine of code to print the shortest distance between all nodes
    print("The shortest distance considering the graph is", distance, "km")
```
