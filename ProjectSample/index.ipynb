```python
import pandas as pd
```

```python
df = pd.read_csv('AviationData.csv', encoding='latin-1')
df.head()
```

```python
df['Location'].fillna(value=df['Location'].mode()[0], inplace=True)
```

```python
# df.isna().sum()
df.tail()
```

```python
df['Country'].fillna(value=df['Country'].mode()[0], inplace=True)
df['Injury.Severity'].fillna(value=df['Injury.Severity'].mode()[0], inplace=True)
df['Aircraft.damage'].fillna(value=df['Aircraft.damage'].mode()[0], inplace=True)
df['Registration.Number'].fillna(value=df['Registration.Number'].mode()[0], inplace=True)
df['Make'].fillna(value=df['Make'].mode()[0], inplace=True)
df['Model'].fillna(value=df['Model'].mode()[0], inplace=True)
df['Amateur.Built'].fillna(value=df['Amateur.Built'].mode()[0], inplace=True)
df['Number.of.Engines'].fillna(value=df['Number.of.Engines'].mode()[0], inplace=True)
df['Engine.Type'].fillna(value=df['Engine.Type'].mode()[0], inplace=True)
df['Purpose.of.flight'].fillna(value=df['Purpose.of.flight'].mode()[0], inplace=True)
df['Weather.Condition'].fillna(value=df['Weather.Condition'].mode()[0], inplace=True)
df['Report.Status'].fillna(value=df['Report.Status'].mode()[0], inplace=True)
df['Total.Uninjured'].fillna(value=df['Total.Uninjured'].mode()[0], inplace=True)
```

```python
df = df.dropna(thresh = 50000, axis = 1)
```
