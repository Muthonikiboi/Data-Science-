```python
# Check difference between append , extend and pop
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.append(list2)
print(list1)

list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)
print(list1)

list1.pop()
print(list1)
```

```python
#Loading and processing a CSV file
import csv

def load_csv(path):
  with open(path, newline = '') as f:
    reader = csv.DictReader(f)
    return list(reader)
    
val = load_csv('./apple_orchard.csv')
val
```

```python
 # SUM
sum = 0
for i in val:
  details = float(i['a`yield'])
  sum += details
print("The SUM is:",sum)
```

```python
#Average
average = sum /len(val)
print("The Average is:",average)
```
