```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data['total_bill'] , bins = 10, color='pink')
plt.title('Distribution of total bill amount')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x = 'day', y='tip' , data=data ,color = 'pink')
plt.title('Tips Distribution per day')
plt.xlabel('Day')
plt.ylabel('Tips')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.regplot(x = 'total_bill', y='sex' , data=data ,color = 'pink')
plt.title('Total bill vs sex')
plt.xlabel('total_bill')
plt.ylabel('sex')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.regplot(x = 'total_bill', y='tip' , scatter_kws = {'alpha':0.5} , data=data  ,color = 'pink')
plt.title('Tips Distribution per day')
plt.xlabel('total_bill')
plt.ylabel('tips')
plt.show()
```
