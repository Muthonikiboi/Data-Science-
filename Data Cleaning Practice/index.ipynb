```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv', encoding = 'latin-1')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['Cabin'].fillna('Unknown', inplace = True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
fill_Mode = df['Embarked'].mode()[0]
```

```python
df['Embarked'].fillna(fill_Mode, inplace = True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
Q1 = df["Age"].quantile(0.25)
Q1
```

```python
Q3 = df["Age"].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

    # Filter outliers
df_Age = df[(df["Age"] >= lower_bound) & (df["Age"] <= upper_bound)]
df_Age
```
